```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_outer(img_src):
    r_img = cv2.imread(img_src)
    g_img = cv2.cvtColor(r_img, cv2.COLOR_BGR2GRAY)
    e_img = cv2.Canny(g_img, 30, 200)
    contours, hierarchy = cv2.findContours(e_img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
    x = -1
    y = -1
    w = 0
    h = 0
    for contour in contours:
        x1, y1, w1, h1 = cv2.boundingRect(contour)
        if x == -1:
            x = x1
        if y == -1:
            y = y1
        if x1 < x:
            x = x1
        if y1 < y:
            y = y1
        if w1 > w:
            w = w1
        if h1 > h:
            h = h1
    r_img_1 = cv2.rectangle(r_img,(x,y),(x+w,y+h),(255,0,0),2)
    plt.imshow(r_img_1)
    plt.title("Outermost Contour")
    plt.grid(False)
    plt.show()
```

```python
draw_outer("edge_images/dl_1.jpeg")
```
